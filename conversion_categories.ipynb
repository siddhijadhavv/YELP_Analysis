```python
import psycopg2
  
# establishing the connection
conn = psycopg2.connect(
   database="Yelp",
    user='postgres',
    password='root',
    host='localhost',
    port= '5432'
)
  
# Creating a cursor object using the cursor()
# method
cursor = conn.cursor()
  

# Display whole table
cursor.execute("SELECT business_id, categories FROM yelp_business")
rows = cursor.fetchall()

business_id = []
categories = []

dict_values = {}

for row in rows:
   dict_values[row[0]] = row[1].split(";")
  
# Commit your changes in the database
conn.commit()
  
# #Closing the connection

dict_values
```

```python
for key, value in dict_values.items():
    for  i in range(0, len(value)):
        business_id.append(key)
```

```python
for key, value in dict_values.items():
    for i in value:
        categories.append(i)
```

```python
conn = psycopg2.connect(
   database="Yelp",
    user='postgres',
    password='root',
    host='localhost',
    port= '5432'
)
  
# Creating a cursor object using the cursor()
# method
cursor = conn.cursor()
create_statement = "CREATE TABLE yelp_categories(business_id varchar(50) not null, categories char(50) not null);"
cursor.execute(create_statement)
print("Table created successfully")

conn.commit()
conn.close()
```

```python
conn = psycopg2.connect(
   database="Yelp",
    user='postgres',
    password='root',
    host='localhost',
    port= '5432'
)
  
# Creating a cursor object using the cursor()
# method
cursor = conn.cursor()
insert_statement = "INSERT INTO yelp_categories values (%s, %s);"
final_list = []
for i, j in zip(business_id, categories):
    final_list.append((i,j))

cursor.executemany(insert_statement, final_list)

conn.commit()
conn.close()
```

```python
import psycopg2
conn = psycopg2.connect(
   database="Yelp",
    user='postgres',
    password='root',
    host='localhost',
    port= '5432'
)
  
# Creating a cursor object using the cursor()
# method
cursor = conn.cursor()
create_statement = "ALTER TABLE yelp_categories ADD FOREIGN KEY (business_id) REFERENCES yelp_business(business_id) on delete cascade;"
cursor.execute(create_statement)
print("Table altered successfully")

conn.commit()
conn.close()
```
